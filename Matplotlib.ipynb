# Matplotlib

```python
from matplotlib import pyplot as plt
```

```python
plt.plot([1,2,3],[4,5,2])
plt.show()
```

### Add title and labels

```python
x = [12,23,14]
y = [12,3,6]

plt.plot(x,y)
plt.title('Example')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

### Adding color to line

```python
x = [12,23,14]
y = [12,3,6]

plt.plot(x,y,'r')
plt.title('Example')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

### Adding legend to the graph

```python
x = [12,23,14]
y = [12,3,6]

plt.plot(x,y,'r',label='line', linewidth = 5)
plt.title('Example')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.show()
```

### Add grid

```python
x = [12,23,14]
y = [12,3,6]

plt.plot(x,y,'r',label='line', linewidth = 5)
plt.title('Example')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.grid(True,color = 'g')
plt.show()
```

### Multiple lines 

```python
x1 = [12,23,14]
y1 = [12,3,6]

x2 = [6,7,8]
y2 = [17,8,29]

plt.plot(x1,y1,'r',label='lineone', linewidth = 5)
plt.plot(x2,y2,'c',label='linetwo', linewidth = 5)

plt.title('Example')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.grid(True,color = 'g')
plt.show()
```

### Change the marker color

```python
from matplotlib import pyplot as plt

x1 = [12,16,20,25,45]
y1 = [12,3,6,14,12]

x2 = [11,15,19,22,26]
y2 = [12,16,14,7,19]

plt.plot(x1,y1,color = 'r',label='lineone',marker = 'o',markerfacecolor='black')
plt.plot(x2,y2,color = 'c',label='linetwo',marker = 'v',markerfacecolor='b')

plt.title('Example')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend(loc = 'upper right')
plt.grid(True,color = 'g')
plt.show()
```

### Changing line style

```python
from matplotlib import pyplot as plt

x1 = [12,16,20,25,45]
y1 = [12,3,6,14,12]

x2 = [11,15,19,22,26]
y2 = [12,16,14,7,19]

plt.plot(x1,y1,color = 'r',label='lineone',marker = 'o',markerfacecolor='black',linestyle = '-.')
plt.plot(x2,y2,color = 'c',label='linetwo',marker = 'v',markerfacecolor='b',linestyle = '-.')

plt.title('Example')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend(loc = 'upper right')
plt.grid(True,color = 'g')
plt.show()
```

# Subplots

```python
#subplot = num of rows, num of col, fig num......
#Eg: subplot(2,1,1) or subplot(211)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10,150)
cos = np.cos(x)
sin = np.sin(x)
plt.subplot(2,1,1)
plt.plot(x,cos,color='red',linewidth=2)
plt.subplot(2,1,2)
plt.plot(x,sin,color='yellow',linewidth=2)
plt.show()
```

```python
x = np.linspace(-10,10,150)
cos = np.cos(x)
sin = np.sin(x)
plt.subplot(1,2,1)
plt.plot(x,cos,color='red',linewidth=2)
plt.subplot(1,2,2)
plt.plot(x,sin,color='yellow',linewidth=2)
plt.show()
```

```python
# dataset-1 
x1 = [89, 43, 36, 36, 95, 10, 66, 34, 38, 20] 
y1 = [21, 46, 3, 35, 67, 95,53, 72, 58, 10] 
# dataset2 
x2 = [26, 29, 48, 64, 6, 5,36, 66, 72, 40] 
  
y2 = [26, 34, 90, 33, 38, 20, 56, 2, 47, 15]
plt.subplot(2,1,1)
plt.title('Scatter plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.scatter(x1,y1,color = 'pink',marker = 's',edgecolor = 'green',linewidth = 2,s = 200)
plt.subplot(2,1,2)
plt.scatter(x2,y2,color = 'yellow',marker = 'o',edgecolor = 'black',linewidth = 2,s = 300)

plt.xlabel('X')
plt.ylabel('Y')
plt.tight_layout()
#plt.legend(['Square','Circle'],loc = 'best',fontsize = 10)
plt.show()
```

# Pie Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12] 
plt.pie(students,labels = langs)
plt.show()
```

### changing figure size

```python
plt.figure(figsize=(10,10))
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12] 
plt.pie(students,labels = langs)
plt.show()
```

```python
plt.figure(figsize=(5,5))
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12] 
plt.pie(students,labels = langs,autopct ='%1.0f%%')
plt.show()
```

```python
 # add_axes() =  left, bottom, width, height
```

```python
fig = plt.figure(figsize=(5,5))
ax = fig.add_axes([0,0,2,1])
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
ax.axis('equal')
ax.pie(students,labels = langs,autopct ='%1.0f%%')
plt.show()
```
