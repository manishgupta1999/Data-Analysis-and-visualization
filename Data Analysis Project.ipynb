# Data Analysis Project

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_literacy = pd.read_csv("literacy.csv")
data_literacy
```

```python
#Sort data according to state
data_literacy = data_literacy.sort_values('State/UT')
data_literacy
```

```python
data_literacy['S.N.'] = np.arange(1,32)
data_literacy
```

```python
data_literacy.set_index('S.N.',inplace=True)
data_literacy
```

```python
covid19 = pd.read_csv("covid19india.csv")
covid19.head()
```

```python
covid19.set_index('S.No.',inplace=True)
```

```python
covid19
```

```python
column_labels = ['State', 'Total confirmed', 'Discharged', 'Death']
covid19.columns = column_labels
covid19
```

```python
all_cases_india = covid19['Total confirmed'].sum()
print("Total case in India", all_cases_india)
```

```python
total_death = covid19['Death'].sum()
print("Total Death in India", total_death)
```

```python
recovered = covid19['Discharged'].sum()
print("Total recovered in India", recovered)
```

# Merging two datasets

```python
data = pd.merge(covid19,data_literacy, how = 'outer', left_on='State',right_on= 'State/UT')
data
```

### Remove unnecessary data

```python
data.drop([17,28],axis = 0 ,inplace = True)
data
```

```python
data.drop(['State/UT','Capitals','Type','Males','Females','Sex Ratio'],axis = 1,inplace=True)
data
```

```python
data['Population'].head()
```

```python
data['Population'].dtype
```

```python
data['Population'] = data['Population'].str.replace(',','')
data['Population'] = data['Population'].astype(int)
data['Population'].head()
```

```python
# calculate percentage cases 
data['%covidCases'] = data['Total confirmed'].div(data['Population'])
data['%covidCases']
```

```python
#Value is too small 
data['%covidCases'] = data['%covidCases']*10000
data['%covidCases'].head()
```

### Plot data using bar plot

```python
fig = plt.figure(figsize= (25,8))
ax = fig.add_axes([0,0,1,1])
ax.set_xlabel('Covid Cases',fontsize=20)
ax.set_ylabel('Percentage',fontsize=20)
ax.bar(data['State'],data['%covidCases'])
plt.show()
```

```python
plt.rcParams['xtick.labelsize'] = 25
plt.rcParams['ytick.labelsize'] = 20
fig = plt.figure(figsize= (25,8))
ax = fig.add_axes([0,0,1,1])
ax.set_xlabel('Covid Cases',fontsize=20)
ax.set_ylabel('Percentage',fontsize=20)
ax.bar(data['State'],data['%covidCases'])
plt.show()
```

```python
plt.rcParams['xtick.labelsize'] = 25
plt.rcParams['ytick.labelsize'] = 20
fig = plt.figure(figsize= (25,8))
ax = fig.add_axes([0,0,1,1])
ax.set_xlabel('Covid Cases',fontsize=20)
ax.set_ylabel('Percentage',fontsize=20)
ax.bar(data['State'],data['%covidCases'])
ax.tick_params(axis='x',rotation=90)
plt.show()
```

### Multiple plots

```python
plt.rcParams['xtick.labelsize'] = 25
plt.rcParams['ytick.labelsize'] = 10
fig = plt.figure(figsize= (30,20))

plt.subplot(2,1,1)
plt.xlabel('State',fontsize=20)
plt.ylabel('% Covid Cases',fontsize=20)
plt.bar(data['State'],data['%covidCases'],color='red')
plt.xticks(data['State'],rotation=90)

plt.subplot(2,1,2)
plt.xlabel('State',fontsize=20)
plt.ylabel('% Population',fontsize=20)
plt.bar(data['State'],data['% Population'],color='green')
plt.xticks(data['State'],rotation=90)
plt.tight_layout()
plt.show()
```

### Combining both plot in one using twinx()

```python
plt.figure(figsize=(20,8))
ax1 = plt.subplot(1,1,1)
w=0.3 #width
x=np.arange(1,32)
plt.xticks(x + w/2,data['State'],rotation='vertical')
ax1.bar(x,data['% Population'],width=w, color='b',align='center')
plt.ylabel('Population',fontsize=20)
ax2=ax1.twinx()
ax2.bar(x+w,data['%covidCases'],width=w, color='r',align='center')
plt.ylabel('Covid Cases',fontsize=20)

plt.show()
```

```python
plt.figure(figsize=(20,8))
ax1 = plt.subplot(1,1,1)
w=0.3 #width
x=np.arange(1,32)
plt.xticks(x + w/2,data['State'],rotation='vertical')
pop=ax1.bar(x,data['% Population'],width=w, color='b',align='center')
plt.ylabel('Population',fontsize=20)
ax2=ax1.twinx()
covidCase=ax2.bar(x+w,data['%covidCases'],width=w, color='r',align='center')
plt.ylabel('Covid Cases',fontsize=20)
plt.legend([pop,covidCase],['%Poplulation','%CovidCases'],fontsize=15)
plt.show()
```
